### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
# Try to import necessary libraries
try:
    import matplotlib.pyplot as plt
    import numpy as np
except ImportError as e:
    print(f"Error importing libraries: {e}")
    # Optionally, you can raise the error to stop further execution
    # raise e

# Create the data for the line plot
x = np.linspace(0, 10, 100)  # x values from 0 to 10
y = x ** 2  # y values as the square of x

# Create the plot
plt.plot(x, y)

# Customize the plot
plt.title('Line Plot: y = x^2')  # Title of the plot
plt.xlabel('X values')  # Label for the x-axis
plt.ylabel('Y values (x^2)')  # Label for the y-axis
plt.grid(True)  # Add gridlines to the plot

# Show the plot
plt.show()
```
