### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
# Try to import the necessary library
try:
    import matplotlib.pyplot as plt
except ImportError as e:
    print(f"Error importing matplotlib: {e}")
    # Optionally, you can raise the error to stop further execution
    # raise e

# Data for the scatter plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create the scatter plot
plt.scatter(x, y, color='blue', label='Data Points')

# Annotate each point with its (x, y) value
for i in range(len(x)):
    plt.text(x[i], y[i], f'({x[i]}, {y[i]})', fontsize=9, ha='right')

# Customize the plot
plt.title('Scatter Plot with Annotations')  # Title of the plot
plt.xlabel('X values')  # Label for the x-axis
plt.ylabel('Y values')  # Label for the y-axis
plt.grid(True)  # Add grid lines for better readability

# Display the legend
plt.legend()

# Show the plot
plt.show()


```
