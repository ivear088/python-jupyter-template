### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
# Try to import the necessary library
# Enhanced error handling for library imports and plot creation

# Step 1: Import libraries with error handling
try:
    import matplotlib.pyplot as plt
except ImportError as e:
    print(f"Error importing matplotlib: {e}")
    raise

# Step 2: Define the data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Step 3: Create a scatter plot with error handling
try:
    # Create the scatter plot
    plt.scatter(x, y, color='blue', label='Data Points')
    
    # Annotate each point with its (x, y) value
    for i, (xi, yi) in enumerate(zip(x, y)):
        plt.text(xi, yi, f'({xi}, {yi})', fontsize=12, ha='right', color='black')

    # Customize the plot
    plt.title('Scatter Plot with Annotations')  # Title
    plt.xlabel('X Values')  # Label for x-axis
    plt.ylabel('Y Values')  # Label for y-axis
    plt.grid(True)  # Show grid
    plt.legend()  # Show legend
    
    # Display the plot
    plt.show()
except Exception as e:
    print(f"Error creating or displaying the plot: {e}")



```
