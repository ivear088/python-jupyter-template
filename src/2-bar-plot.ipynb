### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
# Try to import the necessary library
# Enhanced error handling for library imports and plot creation

# Step 1: Import libraries with error handling
try:
    import matplotlib.pyplot as plt
except ImportError as e:
    print(f"Error importing matplotlib: {e}")
    raise

# Step 2: Define the data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

# Step 3: Create a bar plot with error handling
try:
    # Create the bar plot
    plt.bar(categories, values, color=['red', 'blue', 'green', 'orange'])
    
    # Customize the plot
    plt.title('Bar Plot with Different Colors')  # Title
    plt.xlabel('Categories')  # Label for x-axis
    plt.ylabel('Values')  # Label for y-axis
    plt.grid(True)  # Show grid
    
    # Display the plot
    plt.show()
except Exception as e:
    print(f"Error creating or displaying the plot: {e}")
```
