### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
# Try to import the necessary library
try:
    import matplotlib.pyplot as plt
except ImportError as e:
    print(f"Error importing matplotlib: {e}")
    # Optionally, you can raise the error to stop further execution
    # raise e

# Data for the bar plot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

# Create the bar plot with different colors for each bar
colors = ['red', 'green', 'blue', 'purple']
plt.bar(categories, values, color=colors)

# Customize the plot
plt.title('Bar Plot with Different Colors')  # Title of the plot
plt.xlabel('Categories')  # Label for the x-axis
plt.ylabel('Values')  # Label for the y-axis

# Show the plot
plt.show()
```
