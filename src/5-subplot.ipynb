### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
# Try to import the necessary library
# Enhanced error handling for library imports and subplot creation

# Step 1: Import libraries with error handling
try:
    import matplotlib.pyplot as plt
    import numpy as np
except ImportError as e:
    print(f"Error importing libraries: {e}")
    raise

# Step 2: Create a 2x2 subplot layout with error handling
try:
    # Create the figure and subplots
    fig, axs = plt.subplots(2, 2, figsize=(10, 8))

    # Line plot (Top-left subplot)
    x = np.linspace(0, 10, 100)
    y = np.square(x)
    axs[0, 0].plot(x, y, label='y = x^2', color='b')
    axs[0, 0].set_title('Line Plot')
    axs[0, 0].set_xlabel('X-axis')
    axs[0, 0].set_ylabel('Y-axis')
    axs[0, 0].grid(True)

    # Bar chart (Top-right subplot)
    categories = ['A', 'B', 'C', 'D']
    values = [5, 7, 3, 9]
    axs[0, 1].bar(categories, values, color=['red', 'green', 'blue', 'orange'])
    axs[0, 1].set_title('Bar Chart')
    axs[0, 1].set_xlabel('Categories')
    axs[0, 1].set_ylabel('Values')

    # Scatter plot (Bottom-left subplot)
    x_scatter = [1, 2, 3, 4, 5]
    y_scatter = [2, 4, 6, 8, 10]
    axs[1, 0].scatter(x_scatter, y_scatter, color='purple')
    axs[1, 0].set_title('Scatter Plot')
    axs[1, 0].set_xlabel('X-axis')
    axs[1, 0].set_ylabel('Y-axis')

    # Pie chart (Bottom-right subplot)
    labels = ['Python', 'Java', 'C++', 'JavaScript']
    sizes = [40, 25, 20, 15]
    axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
    axs[1, 1].set_title('Programming Language Usage')

    # Adjust layout
    plt.tight_layout()

    # Show the plots
    plt.show()

except Exception as e:
    print(f"Error creating or displaying subplots: {e}")
```
